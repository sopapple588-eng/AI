```
!pip install Flask pyngrok line-bot-sdk requests --quiet
!pip install google-genai --quiet
from google.colab import userdata

ngrok_authtoken = userdata.get('NGROK_AUTHTOKEN')
line_channel_access_token = userdata.get('LINE_CHANNEL_ACCESS_TOKEN')
line_channel_secret = userdata.get('LINE_CHANNEL_SECRET')
gemini_api_key = userdata.get('GEMINI_API_KEY')
port = 5051
import os
from pyngrok import ngrok
ngrok.kill()
import requests

ngrok.set_auth_token(ngrok_authtoken)
tunnel = ngrok.connect(5051, name="linebot_tunnel")
webhook_url = tunnel.public_url

print(f"Ngrok URL: {webhook_url}")

# è‡ªå‹•æ›´æ–° LINE Webhook URL
def update_line_webhook(webhook_url):
    """ä½¿ç”¨ LINE Messaging API æ›´æ–° Webhook URL"""
    url = "https://api.line.me/v2/bot/channel/webhook/endpoint"
    headers = {
        "Authorization": f"Bearer {line_channel_access_token}",
        "Content-Type": "application/json"
    }
    data = {
        "endpoint": webhook_url
    }

    response = requests.put(url, headers=headers, json=data)

    if response.status_code == 200:
        print(f"âœ… LINE Webhook URL å·²è‡ªå‹•æ›´æ–°ç‚ºï¼š{webhook_url}")
        return True
    else:
        print(f"âŒ æ›´æ–°å¤±æ•—ï¼š{response.status_code} - {response.text}")
        return False

# åŸ·è¡Œæ›´æ–°
update_line_webhook(webhook_url)
Ngrok URL: https://unnecessitated-kendall-gogetting.ngrok-free.dev
âœ… LINE Webhook URL å·²è‡ªå‹•æ›´æ–°ç‚ºï¼šhttps://unnecessitated-kendall-gogetting.ngrok-free.dev
True
from google import genai
from google.genai.types import Tool, GenerateContentConfig, GoogleSearch

# === åˆå§‹åŒ– Google Gemini ===
client = genai.Client(api_key=gemini_api_key)

google_search_tool = Tool(
   google_search=GoogleSearch()
)

chat = client.chats.create(
    model="gemini-2.5-flash",
    config=GenerateContentConfig(
        system_instruction="ä½ æ˜¯ä¸€å€‹ä¸­æ–‡çš„AIåŠ©æ‰‹ï¼Œè«‹ç”¨ç¹é«”ä¸­æ–‡å›ç­”",
        tools=[google_search_tool],
        response_modalities=["TEXT"],
    )
)
def stateful_query(payload):
    response = chat.send_message(message=payload)
    return response.text
result = stateful_query("ç°¡ä»‹æ˜æ–°ç§‘æŠ€å¤§å­¸")
print(result)
æ˜æ–°ç§‘æŠ€å¤§å­¸ï¼ˆMinghsin University of Science and Technologyï¼Œç°¡ç¨±æ˜æ–°ç§‘å¤§ï¼‰æ˜¯ä¸€æ‰€ä½æ–¼å°ç£æ–°ç«¹ç¸£æ–°è±é„‰çš„ç§ç«‹ç§‘æŠ€å¤§å­¸ã€‚å­¸æ ¡çš„å‰µç«‹å¯è¿½æº¯è‡³1966å¹´3æœˆï¼Œç•¶æ™‚ä»¥ã€Œæ˜æ–°å·¥æ¥­å°ˆç§‘å­¸æ ¡ã€ç«‹æ¡ˆï¼Œä¸¦æ–¼åŒå¹´3æœˆ1æ—¥æ­£å¼æ‹›æ”¶æ–°ç”Ÿï¼Œè¨­æœ‰æ©Ÿæ¢°ã€åœŸæœ¨ã€å·¥æ¥­ç®¡ç†ä¸‰ç§‘äº”å¹´åˆ¶èª²ç¨‹ã€‚

å­¸æ ¡çš„ç™¼å±•æ­·ç¨‹å¦‚ä¸‹ï¼š
*   1966å¹´ï¼šæ˜æ–°å·¥æ¥­å°ˆç§‘å­¸æ ¡å‰µæ ¡ã€‚
*   1993å¹´ï¼šæ›´åç‚ºæ˜æ–°å·¥å•†å°ˆç§‘å­¸æ ¡ã€‚
*   1997å¹´7æœˆï¼šå¥‰æ•™è‚²éƒ¨æ ¸å‡†æ”¹åˆ¶ç‚ºã€Œæ˜æ–°æŠ€è¡“å­¸é™¢ã€ï¼Œä¸¦é™„è¨­å°ˆç§‘éƒ¨ã€‚
*   2002å¹´8æœˆï¼šå¥‰æ•™è‚²éƒ¨æ ¸å‡†å„ªå…ˆå‡æ ¼ç‚ºã€Œæ˜æ–°ç§‘æŠ€å¤§å­¸ã€ï¼Œä¸¦æ²¿ç”¨æ­¤æ ¡åã€‚
*   2018å¹´12æœˆï¼šå¥‰æ•™è‚²éƒ¨æ ¸å‡†æ›´åç‚ºã€Œæ˜æ–°å­¸æ ¡è²¡åœ˜æ³•äººæ˜æ–°ç§‘æŠ€å¤§å­¸ã€ã€‚

æ˜æ–°ç§‘æŠ€å¤§å­¸çš„æ ¡åã€Œæ˜æ–°ã€å–è‡ªã€Šå¤§å­¸ã€‹ã€Œåœ¨æ˜æ˜å¾·ï¼Œåœ¨æ–°æ°‘ï¼Œåœ¨æ­¢æ–¼è‡³å–„ã€çš„ç²¾ç¾©ï¼Œæ—¨åœ¨é—¡æšäººé¡èˆ‡ç”Ÿä¿±ä¾†çš„å¾·æ€§èˆ‡æƒ…æ“ï¼ŒåŸ¹è‚²å­¸å­å…·å‚™é«˜å°šå“å¾·ã€å°ˆæ¥­å­¸å•èˆ‡å„ªè‰¯æŠ€è¡“ï¼Œä»¥é”æˆã€Œä¸€æµç”¢æ¥­å¤§å­¸ã€çš„ç™¼å±•é¡˜æ™¯ï¼Œä¸¦åŸ¹è‚²å‡ºã€Œè·¨åŸŸæ•´åˆã€å‹™å¯¦å‰µæ–°ã€å…¨äººå­¸ç¿’ã€çš„å°ˆæ¥­äººæ‰ï¼Œç¬¦åˆæ ¡è¨“ã€Œå …æ¯…ã€æ±‚æ–°ã€å‰µé€ ã€çš„ç²¾ç¥ã€‚

ç›®å‰ï¼Œæ˜æ–°ç§‘æŠ€å¤§å­¸è¨­æœ‰å…­å€‹å­¸é™¢ï¼ŒåŒ…æ‹¬åŠå°é«”å­¸é™¢ã€å·¥ç¨‹å­¸é™¢ã€ç®¡ç†å­¸é™¢ã€æ°‘ç”Ÿå­¸é™¢ã€äººæ–‡èˆ‡è¨­è¨ˆå­¸é™¢ä»¥åŠå…±åŒæ•™è‚²å­¸é™¢ã€‚å­¸æ ¡æ¶µæ‹¬20å€‹å­¸ç³»ã€2å€‹å­¸ä½å­¸ç¨‹ï¼ˆå«1å€‹åšå£«å­¸ä½å­¸ç¨‹ï¼‰ä»¥åŠ11å€‹ç¢©å£«ç­ã€‚æˆªè‡³ç›®å‰ï¼Œå­¸ç”Ÿç¸½æ•¸ï¼ˆå«æ—¥é–“éƒ¨èˆ‡é€²ä¿®éƒ¨ï¼‰ç´„æœ‰ä¸€è¬ä¸‰åƒé¤˜äººï¼Œå°ˆä»»æ•™å¸«ç´„ä¸‰ç™¾å…«åé¤˜äººï¼Œç•¢æ¥­æ ¡å‹å·²é€¾ä¹è¬é¤˜äººã€‚2022å¹´10æœˆï¼Œå­¸æ ¡ç²æ•™è‚²éƒ¨æ ¸å‡†é€šéã€ŒåŠå°é«”ç§‘æŠ€åšå£«å­¸ä½å­¸ç¨‹ã€ï¼Œé€™æ˜¯æ˜æ–°ç§‘å¤§æˆç«‹äº”åå…­å¹´ä¾†çš„ç¬¬ä¸€å€‹åšå£«ç­ã€‚
result2 = stateful_query("æ ¡é•·æ˜¯èª°ï¼Ÿ")
print(result2)
æ˜æ–°ç§‘æŠ€å¤§å­¸ç¾ä»»æ ¡é•·æ˜¯**å‘‚æ˜å³¯**æ•™æˆã€‚ä»–æ–¼2025å¹´2æœˆ1æ—¥æ­£å¼ä¸Šä»»ï¼Œæˆç‚ºæ˜æ–°ç§‘å¤§ç¬¬11ä»»æ ¡é•·ã€‚

å‘‚æ˜å³¯æ•™æˆåœ¨æ¥ä»»æ ¡é•·ä¸€è·å‰ï¼Œæ›¾æ“”ä»»æ˜æ–°ç§‘æŠ€å¤§å­¸é›»å­å·¥ç¨‹ç³»æ•™æˆã€ç”¢å­¸åˆä½œè™•è™•é•·ã€å·¥å­¸é™¢é™¢é•·ã€åŠå°é«”å­¸é™¢é™¢é•·ç­‰è·å‹™ï¼Œæ“æœ‰è±å¯Œçš„åŠå°é«”æ•™è‚²åŠç”¢æ¥­å¯¦å‹™ç¶“é©—ã€‚ ä»–å°‡ä»¥ã€Œå››å¤§æ ¸å¿ƒæ¨¡çµ„ã€ä½œç‚ºæ ¡å‹™æ²»ç†ç­–ç•¥ï¼Œå¸¶é ˜æ˜æ–°ç§‘å¤§é–‹å‰µæŠ€è·æ•™è‚²æ–°å±€ï¼Œç›®æ¨™æ˜¯å°‡æ˜æ–°ç§‘å¤§æ‰“é€ æˆç‚ºã€Œå…·åœ‹éš›é­…åŠ›çš„ç”¢æ¥­ç§‘æŠ€å¤§å­¸ã€ã€‚

åœ¨å‘‚æ˜å³¯æ ¡é•·ä¸Šä»»ä¹‹å‰ï¼Œæ˜æ–°ç§‘æŠ€å¤§å­¸çš„æ ¡é•·æ˜¯æ—å•Ÿç‘ã€‚
from flask import Flask, request, abort
import logging
import os
import time
from google.genai import types

from linebot.v3 import (
    WebhookHandler
)
from linebot.v3.exceptions import (
    InvalidSignatureError
)
from linebot.v3.messaging import (
    Configuration,
    ApiClient,
    MessagingApi,
    MessagingApiBlob,
    ReplyMessageRequest,
    TextMessage
)
from linebot.v3.webhooks import (
    MessageEvent,
    TextMessageContent,
    FileMessageContent
)

app = Flask(__name__)

logging.basicConfig(
    level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s"
)
app.logger.setLevel(logging.INFO)

configuration = Configuration(access_token=line_channel_access_token)
handler = WebhookHandler(line_channel_secret)

# å„²å­˜æª”æ¡ˆçš„ç›®éŒ„
UPLOAD_DIR = "/content/uploaded_files"
os.makedirs(UPLOAD_DIR, exist_ok=True)

# å„²å­˜æ¯å€‹ä½¿ç”¨è€…çš„å°è©± session å’Œä¸Šå‚³çš„æª”æ¡ˆ
user_sessions = {}  # {user_id: {"chat": chat_object, "uploaded_file": gemini_file}}

def get_user_session(user_id):
    """å–å¾—æˆ–å»ºç«‹ä½¿ç”¨è€…çš„å°è©± session"""
    if user_id not in user_sessions:
        # å»ºç«‹æ–°çš„å°è©± session
        new_chat = client.chats.create(
            model="gemini-2.5-flash",
            config=GenerateContentConfig(
                system_instruction="ä½ æ˜¯ä¸€å€‹ä¸­æ–‡çš„AIåŠ©æ‰‹ï¼Œè«‹ç”¨ç¹é«”ä¸­æ–‡å›ç­”ã€‚å¦‚æœä½¿ç”¨è€…æœ‰æä¾›åƒè€ƒæ–‡ä»¶ï¼Œè«‹æ ¹æ“šæ–‡ä»¶å…§å®¹å›ç­”å•é¡Œã€‚",
                tools=[google_search_tool],
                response_modalities=["TEXT"],
            )
        )
        user_sessions[user_id] = {
            "chat": new_chat,
            "uploaded_file": None
        }
    return user_sessions[user_id]

def download_line_file(message_id, file_name):
    """å¾ LINE ä¸‹è¼‰ä½¿ç”¨è€…ä¸Šå‚³çš„æª”æ¡ˆ"""
    with ApiClient(configuration) as api_client:
        line_bot_blob_api = MessagingApiBlob(api_client)
        file_content = line_bot_blob_api.get_message_content(message_id)

        file_path = os.path.join(UPLOAD_DIR, file_name)

        with open(file_path, 'wb') as f:
            f.write(file_content)

        return file_path

def upload_file_to_gemini(file_path):
    """ä¸Šå‚³æª”æ¡ˆåˆ° Gemini Files API"""
    uploaded_file = client.files.upload(
        file=file_path,
        config={'display_name': os.path.basename(file_path)}
    )

    # ç­‰å¾…æª”æ¡ˆè™•ç†å®Œæˆ
    while uploaded_file.state.name == "PROCESSING":
        print("æª”æ¡ˆè™•ç†ä¸­...")
        time.sleep(1)
        uploaded_file = client.files.get(name=uploaded_file.name)

    if uploaded_file.state.name == "FAILED":
        raise Exception("æª”æ¡ˆä¸Šå‚³è™•ç†å¤±æ•—")

    return uploaded_file

def query_with_rag(user_id, question):
    """ä½¿ç”¨ RAG æ¨¡å¼å›ç­”å•é¡Œ"""
    session = get_user_session(user_id)
    uploaded_file = session["uploaded_file"]

    if uploaded_file:
        # æœ‰ä¸Šå‚³æª”æ¡ˆï¼Œä½¿ç”¨ RAG æ¨¡å¼
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=[
                types.Content(
                    role="user",
                    parts=[
                        types.Part.from_uri(
                            file_uri=uploaded_file.uri,
                            mime_type=uploaded_file.mime_type
                        ),
                        types.Part.from_text(text=f"è«‹æ ¹æ“šä¸Šè¿°æä¾›çš„æª”æ¡ˆå…§å®¹ï¼Œç”¨ç¹é«”ä¸­æ–‡å›ç­”é€™å€‹å•é¡Œï¼š{question}")
                    ]
                )
            ]
        )
        return response.text
    else:
        # æ²’æœ‰ä¸Šå‚³æª”æ¡ˆï¼Œä½¿ç”¨ä¸€èˆ¬å¤šè¼ªå°è©±
        response = session["chat"].send_message(message=question)
        return response.text

@app.route("/", methods=['POST'])
def callback():
    signature = request.headers['X-Line-Signature']
    body = request.get_data(as_text=True)
    print("BODY: ", body)
    app.logger.info("Request body: " + body)

    try:
        handler.handle(body, signature)
    except InvalidSignatureError:
        app.logger.info("Invalid signature.")
        abort(400)

    return 'OK'

@handler.add(MessageEvent, message=TextMessageContent)
def handle_message(event):
    """è™•ç†æ–‡å­—è¨Šæ¯"""
    text = event.message.text
    user_id = event.source.user_id

    with ApiClient(configuration) as api_client:
        line_bot_api = MessagingApi(api_client)

        if text.startswith('AI '):
            prompt = text[3:]
            try:
                # ä½¿ç”¨ RAG æˆ–ä¸€èˆ¬å°è©±
                reply_text = query_with_rag(user_id, prompt)

                # æª¢æŸ¥æ˜¯å¦æœ‰ä¸Šå‚³æª”æ¡ˆï¼ŒåŠ ä¸Šæç¤º
                session = get_user_session(user_id)
                if session["uploaded_file"]:
                    reply_text = f"ğŸ“„ [RAG æ¨¡å¼]\n\n{reply_text}"

                line_bot_api.reply_message_with_http_info(
                    ReplyMessageRequest(
                        reply_token=event.reply_token,
                        messages=[TextMessage(text=reply_text)]
                    )
                )
            except Exception as e:
                line_bot_api.reply_message_with_http_info(
                    ReplyMessageRequest(
                        reply_token=event.reply_token,
                        messages=[TextMessage(text=f"âŒ ç™¼ç”ŸéŒ¯èª¤ï¼š{str(e)}")]
                    )
                )
        elif text == "æ¸…é™¤æ–‡ä»¶":
            # æ¸…é™¤ä½¿ç”¨è€…ä¸Šå‚³çš„æª”æ¡ˆ
            session = get_user_session(user_id)
            session["uploaded_file"] = None
            line_bot_api.reply_message_with_http_info(
                ReplyMessageRequest(
                    reply_token=event.reply_token,
                    messages=[TextMessage(text="âœ… å·²æ¸…é™¤ä¸Šå‚³çš„æ–‡ä»¶ï¼Œæ¢å¾©ä¸€èˆ¬å°è©±æ¨¡å¼ã€‚")]
                )
            )
        else:
            line_bot_api.reply_message_with_http_info(
                ReplyMessageRequest(
                    reply_token=event.reply_token,
                    messages=[TextMessage(text="è«‹è¼¸å…¥ã€ŒAI å•é¡Œã€ä¾†é–‹å§‹å°è©±\næˆ–ä¸Šå‚³ TXT/PDF æª”æ¡ˆå•Ÿç”¨ RAG æ¨¡å¼")]
                )
            )

@handler.add(MessageEvent, message=FileMessageContent)
def handle_file_message(event):
    """è™•ç†ä½¿ç”¨è€…ä¸Šå‚³çš„æª”æ¡ˆ"""
    user_id = event.source.user_id
    file_name = event.message.file_name

    with ApiClient(configuration) as api_client:
        line_bot_api = MessagingApi(api_client)

        # æª¢æŸ¥æª”æ¡ˆé¡å‹
        if not (file_name.endswith('.txt') or file_name.endswith('.pdf')):
            line_bot_api.reply_message_with_http_info(
                ReplyMessageRequest(
                    reply_token=event.reply_token,
                    messages=[TextMessage(text="âš ï¸ ç›®å‰åªæ”¯æ´ TXT æˆ– PDF æª”æ¡ˆ")]
                )
            )
            return

        try:
            # ä¸‹è¼‰æª”æ¡ˆ
            file_path = download_line_file(event.message.id, file_name)
            print(f"æª”æ¡ˆå·²ä¸‹è¼‰ï¼š{file_path}")

            # ä¸Šå‚³åˆ° Gemini
            uploaded_file = upload_file_to_gemini(file_path)
            print(f"æª”æ¡ˆå·²ä¸Šå‚³åˆ° Geminiï¼š{uploaded_file.uri}")

            # å„²å­˜åˆ°ä½¿ç”¨è€… session
            session = get_user_session(user_id)
            session["uploaded_file"] = uploaded_file

            line_bot_api.reply_message_with_http_info(
                ReplyMessageRequest(
                    reply_token=event.reply_token,
                    messages=[TextMessage(text=f"âœ… æª”æ¡ˆã€Œ{file_name}ã€ä¸Šå‚³æˆåŠŸï¼\n\nç¾åœ¨æ‚¨å¯ä»¥è¼¸å…¥ã€ŒAI å•é¡Œã€ä¾†è©¢å•é—œæ–¼é€™ä»½æ–‡ä»¶çš„å•é¡Œã€‚\n\nè¼¸å…¥ã€Œæ¸…é™¤æ–‡ä»¶ã€å¯æ¢å¾©ä¸€èˆ¬å°è©±æ¨¡å¼ã€‚")]
                )
            )
        except Exception as e:
            line_bot_api.reply_message_with_http_info(
                ReplyMessageRequest(
                    reply_token=event.reply_token,
                    messages=[TextMessage(text=f"âŒ æª”æ¡ˆè™•ç†å¤±æ•—ï¼š{str(e)}")]
                )
            )

if __name__ == "__main__":
    app.run(port=port)
```
